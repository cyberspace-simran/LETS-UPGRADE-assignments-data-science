```python
CP = float(input(" Please Enter the Cost Price : "))

SP = float(input(" Please Enter the Sale Price : "))

if(CP > SP):
    amount = CP - SP
    print("Loss = {0}".format(amount))

elif(SP > CP):
    amount = SP - CP
    print("Profit = {0}".format(amount))
else:
    print("NEITHER")
```
