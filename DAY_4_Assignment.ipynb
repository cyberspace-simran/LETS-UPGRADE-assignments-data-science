### Questions 2:
How to create a series from a numpy array?


```python
import numpy as np

import pandas as pd

arr = np.array([1, 24, 23, 4, 12, 14, 15, 28, 29, 26, 5, 24, 43, 24, 54])
print("Array :\n", arr)

print("Panda Series:\n", pd.Series(arr))
```

### Questions 3:
How to convert the index of a series into a column of a dataframe?

```python
import pandas as pd

my_series = pd.Series([1, 24, 23, 4, 12, 14, 15, 28, 29, 26, 5, 24, 43, 24, 54])
print("Series is :", my_series)

print("convert the index of a series into a column of a dataframe :", my_series.to_frame().reset_index())
```

### Questions 5:
Which country origin cars are a part of this dataset?

```python
import seaborn as sb
import pandas as pd
dataset_of_mpg = sb.load_dataset('mpg')
print(dataset_of_mpg)

df = pd.DataFrame(dataset_of_mpg)
print("country origin cars :", df.origin.unique())
```

### Questions 6:
Extract the part of the dataframe which contains cars belonging to 'usa'

```python
import seaborn as sb
import pandas as pd

dataset_of_mpg = sb.load_dataset('mpg')
print(dataset_of_mpg)

df = pd.DataFrame(dataset_of_mpg)

print(" the dataframe which contains cars belonging to 'usa' :", df[df['origin'].str.contains("usa")])
```
